```python
import os
from dotenv import load_dotenv
load_dotenv()
import requests

def githubRequestAuthorized(resource):
    authToken = os.getenv("githubTokenApi")
    if not authToken:
        raise ValueError("You need a token")
    else:
        print("We have a github token {}".format(authToken[0:4]))
    headers = {
        "Authorization": "token {}".format(authToken)
    }
    url = "https://api.github.com{}".format(resource)
    print("Requesting authorized {}".format(url))
    res = requests.get(url, headers=headers)
    return res


data = githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents").json()

```

```python
display(data[5])
path_list = [e["path"] for e in data]
print(path_list)
print(len(path_list))
```

```python
data = []
for e in path_list:
    url = "/repos/ironhack-datalabs/scavenger/contents/{}/".format(e)
    data.append(githubRequestAuthorized(url).json())

```
