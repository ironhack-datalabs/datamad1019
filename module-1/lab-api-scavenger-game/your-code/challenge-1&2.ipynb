```python

## CHALLENGE 1

import os
from dotenv import load_dotenv
load_dotenv()
import requests

def githubRequestAuthorized(resource):
    authToken = os.getenv("githubTokenApi")
    if not authToken:
        raise ValueError("You need a token")
    else:
        print("We have a github token {}".format(authToken[0:4]))
    headers = {
        "Authorization": "token {}".format(authToken)
    }
    url = "https://api.github.com{}".format(resource)
    print("Requesting authorized {}".format(url))
    res = requests.get(url, headers=headers)
    return res
```

```python
data = githubRequestAuthorized("/repos/ironhack-datalabs/datamad1019/forks").json()
```

```python
display(data)
```

```python
languages = []

for e in data:
    fullName = e["full_name"]
    getting = "/repos/{}/languages".format(fullName)
    languages.append(githubRequestAuthorized(getting).json())

print(languages)
```

```python
list(set([j for e in languages for j in list(e)]))
```

```python
## CHALLENGE 2: we will check commits in the last 24 hours
users = [e["full_name"] for e in data]
branches= []
for e in users:
        getting = "/repos/{}/branches".format(e)
        branches.append(githubRequestAuthorized(getting).json())

```

```python
len(branches)
len(users)
a = list(zip(users, branches))
user_branch = [[e[0], j["name"]] for e in a for j in e[1]]
print(user_branch)
```

```python
commits = []
for e in user_branch:
    getting = "/repos/{}/commits?since=2019-11-05&sha={}&per_page=100".format(e[0], e[1])
    commits.append(githubRequestAuthorized(getting).json())
```

```python
commits = [e for e in commits if e]
len(commits)

#39 commits in last 24 hours
```
