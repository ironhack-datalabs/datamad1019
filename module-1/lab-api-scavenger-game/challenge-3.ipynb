```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
def githubRequestAuthorized(resource):
    authToken = os.getenv("GITHUB_API_TOKEN")
    if not authToken:
        raise ValueError("NECESITAS UN TOKEN")
    else:
        print("We have a github token: ", authToken[0:4])
    headers = {
        "Authorization": "token {}".format(authToken)
    }
    url = "https://api.github.com{}".format(resource)
    print("Requesting authorized {}".format(url))
    res = requests.get(url, headers=headers)
    return res
```

```python
data = githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents").json()
```

```python
data
```

```python
lista_path=[]
for dicc in data:
    for x, y in dicc.items():
        if x == 'path':
            lista_path.append(y)
            
paths = lista_path[1:]
print(lista_path)
print(paths)
```

```python
files=[]
for pt in paths:
    files.append(githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents/{}".format(pt)).json())
```

```python
lista_pathf=[]
for lista in files:
    for dicc in lista:
        for a, b in dicc.items():
            if a == 'path':
                lista_pathf.append(b)
            
print(lista_pathf)
```

```python
import re
```

```python
secret = re.findall("[0-9]+\/\.[0-9]+\.[a-z]+", str(lista_pathf))
secret_order = sorted(secret, key=lambda s: (s[9], s[10]))
```

```python
open_secret=[]
for sc in secret_order:
    open_secret.append(githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents/{}".format(sc)).json())
print(open_secret)
```

```python
frase=[]
for i in range(0,len(open_secret)):
    frase.append(open_secret[i]["content"])
print(frase)
```
