```python
import os
import re
import requests
from dotenv import load_dotenv
load_dotenv()
```

```python
def githubRequestAuthorized(resource):
    authToken = os.getenv("GITHUB_API_TOKEN")
    if not authToken:
        raise ValueError("NECESITAS UN TOKEN")
    else:
        print("We have a github token: ", authToken[0:4])
    headers = {
        "Authorization": "token {}".format(authToken)
    }
    url = "https://api.github.com{}".format(resource)
    print("Requesting authorized {}".format(url))
    res = requests.get(url, headers=headers)
    return res
```

```python
data = githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents").json()
```

```python
data
```

```python
lst_folders = []
for dic in data:
    for k,v in dic.items():
        if k == "name":
            lst_folders.append(v)
            
lst_folders = lst_folders[1:]
print(lst_folders)
```

```python
paths = []
for path in lst_folders:
    paths.append(githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents/{}".format(path)).json())
paths
```

```python
paths[0][0]["name"]
```

```python
lst_files = []
for lista in paths:
    for dic in lista:
        for k,v in dic.items():
            if k == "path":
                lst_files.append(v)
print(lst_files)
```

```python
# Convierto la lista a un string porque sino no puedo hacer regex
lst_files = str(lst_files)
lst_files
```

```python
# Me quedo solo con los archivos que continenen scavengerhunt porque son los que nos interesan
filtro = re.findall("[0-9]+\/\.[0-9]+\.[a-z]+", lst_files)
filtro
```

```python
# Ordeno por archivo
filtro_sorted = sorted(filtro, key=lambda s: (s[9], s[10]))
filtro_sorted
```

```python
secret = []
for secreto in filtro_sorted:
    secret.append(githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents/{}".format(secreto)).json())
secret
```

```python
# Acceso al archivo donde se encuentran las palabras
secret_file = []
for s in filtro_sorted:
    secret_file.append(githubRequestAuthorized("/repos/ironhack-datalabs/scavenger/contents/{}".format(s)).json())
secret_file
```

```python
# Contenido codificado
frase = []
for i in range(1, len(secret_file)):
    frase.append(secret_file[i]["content"])
print(frase)
len(frase)
```

```python
# De-CODE base64
import base64
```

```python
frase_limpia = []
for palabra in frase:
    frase_limpia.append(base64.b64decode(palabra))
print(frase_limpia)
```

```python
# Quito (b)
limpia = re.sub("b+'", "", str(frase_limpia))
# Quito (\\n)
limpia1 = re.sub("\\\\n", "", str(limpia))
# Quito (')
limpia2 = re.sub("\'", "", str(limpia1))
```

```python
# Quito (,) porque he intentado hacer join y en vez de juntarme por las comas, me hace un split de cada palabra por comas...
frase_final = re.sub("\,", "", str(limpia2))
frase_final
```
