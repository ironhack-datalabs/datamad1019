```python
import requests
import json

def getPokemon(number):
    url = "https://pokeapi.co/api/v2/pokemon/{}".format(number)
    res = requests.get(url)
    #print(res)
    # requests.get works asking something to an url, you allocate info extracted in variable res
    data = res.json()
    # res.json automatically modifies info in res to be viewed as json
    #print(data)
    return {
        "name": data["name"],
        "image": data["sprites"]["front_default"]
    }

getPokemon(25)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

# https://developer.github.com/v3/
def githubRequestAuthorized(resource):
    authToken = os.getenv("token")
    if not authToken:
        raise ValueError("NECESITAS UN TOKEN")
    else:
        print("We have a github token: ", authToken[0:2])
    headers = {
        "Authorization": "token {}".format(authToken)
    }
    url = "https://api.github.com{}".format(resource)
    #print("Requesting authorized {}".format(url))
    res = requests.get(url, headers=headers)
    return res
    # requests.get corresponds to one request
```
