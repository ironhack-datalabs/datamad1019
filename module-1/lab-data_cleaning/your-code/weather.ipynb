## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import pymysql 
import getpass
from sqlalchemy import create_engine
```

```python
df=pd.read_csv("../weather-raw.csv")
display(df)
```

```python
melted_cols = df.columns[4:-1]
x=df.melt(id_vars=["year","month",'element'],value_vars=melted_cols)
x.columns
y=x.groupby(["year","month","element"]).mean()
y_max=y.groupby(["month"]).max()
y_min=y.groupby(["month"]).min()
y_max=y_max.rename(columns={"value":"tmax"})
y_min=y_min.rename(columns={"value":"tmin"})
df_merge=pd.merge(y_max,y_min,on="month")
df2=df[["year","month"]]
df2_merge=pd.merge(df2,df_merge,on="month")
d=df2_merge.groupby(["month","tmax","tmin"]).first()
d
```
