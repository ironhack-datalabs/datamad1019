## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import pymysql
import sqlalchemy as db
from sqlalchemy import create_engine
import getpass
```

```python
weather = pd.read_csv('../weather-raw.csv')
weather.head()
```

```python
melted_cols = weather.columns[4:-1]
x = weather.melt(id_vars=["year","month",'element'],value_vars=melted_cols)
temp = x.groupby('month').value.agg({'min','max'})
x = pd.merge(weather[['year','month']],temp, on = 'month')
x
```
