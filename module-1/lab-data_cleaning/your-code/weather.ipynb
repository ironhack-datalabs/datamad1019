This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
df = pd.read_csv('../weather-raw.csv')
```

```python
df.head()
```

Con las siguientes alteraciones se pretende conseguir una tabla con columnas: año, mes, día, temperatura máxima y mínima con el fin de crear una estructura con la que trabajar de forma más fácil

```python
#Tablas independientes que filtran el data set por tipo de temperatura: máxima y mínima
tmax = df[df['element'] == 'tmax']
tmin = df[df['element']== 'tmin']
```

```python
#Tablas en las que los días se trasponen para pasar a ser filas

tmax_2 = pd.melt(tmax, id_vars = ['id','year', 'month', 'element'], var_name = 'day', value_name = 'tmax')
tmin_2 = pd.melt(tmin, id_vars = ['id','year', 'month', 'element'], var_name = 'day', value_name = 'tmin')
tmax_2.head()
```

```python
#Selección de columnas

tmax_3 = tmax_2[['year', 'month', 'day', 'tmax']]
tmin_3 = tmin_2[['year', 'month', 'day', 'tmin']]
```

```python
#Combinación de tablas 

merged = pd.merge(tmax_3, tmin_3)
```

```python
# Filtro para deshacerse de los valores nulos 

filtered = merged[(merged.tmax.notnull()) &(merged.tmin.notnull())]
```

```python
# Transformamos el formato de las etiquetas de los días en números para poder ordenar posteriormente

filtered['day']= filtered['day'].str.replace('d','')
filtered['day'] = filtered['day'].astype('int64')
print(filtered.dtypes)
filtered.head()
```

## Resultado

```python
sorted = filtered.sort_values(by=['month', 'day'], ascending=[True, True])
sorted.head()
```

```python
sorted[['tmax', 'tmin']].describe()
```

#### Posibles análisis

```python
#Cálculo de la temperatura media (max y min) por mes

avg_month = filtered.groupby('month', as_index=False).agg({'tmax':'mean','tmin':'mean'})

avg_month.head()
```

```python
plt.plot( 'month', 'tmax', data=avg_month, color = 'red', linewidth=2)
plt.plot( 'month', 'tmin', data=avg_month, color='blue', linewidth=2)
plt.legend()
plt.title('Evolución mensual de la temperatura media')
```

```python
#Pivot table: media, mínimo, máx por mes 
filtered.pivot_table(index=['year','month'], values = ['tmax', 'tmin'], aggfunc = [np.mean, np.max, np.min])
```
